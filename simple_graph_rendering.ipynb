# 線グラフ

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pyplot as ppl
import matplotlib as mpl
%matplotlib inline
```

```python
np.random.seed(12) # シード値同じだと同じ乱数を発生する
for i in range(8):
    x = np.arange(1000)
    y = np.random.randn(1000).cumsum() # cumulative sum: 累積和
    plt.plot(x, y, label=str(i))
plt.legend()
```

```python
import seaborn as sns # matplotのラッパー。なので、seaborn でできることは matplot でもできる。
%matplotlib inline
```

```python
x1 = np.random.randn(80)
x2 = np.random.randn(80)
x3 = x1 * x2
y1 = .5 + 2 * x1 - x2 + 2.5 * x3 + 3 * np.random.randn(80)
y2 = .5 + 2 * x1 - x2 + 2.5 * np.random.randn(80)
y3 = y2 + np.random.randn(80)
sns.violinplot([x1, x2, x3], scale='area')
```
