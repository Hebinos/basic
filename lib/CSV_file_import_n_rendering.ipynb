# CSVファイルをインポートして色々見てみる

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
player = 'Roger Federer'
filename = "data/{name}.csv".format(name=player.replace(' ', '-'))
df = pd.read_csv(filename)
```

```python
df['win'] = df['winner'] == player
df['win'].tail()
```

```python
hoge = pd.DataFrame([1, 2, 3, 4, 5])
hoge.mean()
```

```python
date = pd.to_datetime(df['start date'])
```

```python
df['dblfaults'] = (df['player1 double faults']) / (df['player1 total points total'])
```

```python
df['dblfaults'].tail()
```

```python
df['dblfaults'].describe()
```

```python
df.groupby('surface')['win'].mean()
```

```python
gb = df.groupby('year')
```

# 散布図と線グラフで表示する

```python
plt.plot_date(date, df['dblfaults'], alpha=.25, lw=0)
plt.plot_date(pd.to_datetime(gb['start date'].max()), gb['dblfaults'].mean(), '-', lw=3)
plt.xlabel('Year')
plt.ylabel('Proportion of double faults per match.')
```
